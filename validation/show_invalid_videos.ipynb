```python
import os
import pandas as pd
import numpy as np
from os.path import join as pjoin
from matplotlib import pyplot as plt
import matplotlib.font_manager as font_manager 
```

```python
# define path
support_path = './support_files'
csv_path = pjoin(support_path, 'invalid_videos.csv')
# change to path of current file
os.chdir(os.path.abspath(''))
# load csv
invalid_csv = pd.read_csv(csv_path)
```

```python
# plot the distribution of number of unrecognized samples
font = font_manager.FontProperties(fname=pjoin(support_path, 'arial.ttf'), size=20, weight='bold')
# obtain value
label = invalid_csv['class_name'].tolist()
value = invalid_csv['num'].tolist()
# define plot detail
fig,ax = plt.subplots(figsize=(25,50))
ax.barh(label, value, color='#0A81AB')
# set ax info
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(4)
ax.spines['left'].set_linewidth(4)
ax.set_xlim([0, 20])
ax.set_ylim([-1, 180])
ax.set_yticks(np.arange(180))
ax.set_xticks(np.linspace(0,20,5))
ax.set_yticklabels(label)
ax.tick_params(axis='x', labelsize=10, width=4, length=10, direction="in", pad=10)
ax.tick_params(axis='y', labelsize=10, width=4, length=0.1, direction="in", pad=15)
ax.invert_yaxis()  # labels read top-to-bottom
# set tick info
plt.yticks(fontproperties=font, size=16)
plt.xticks(fontproperties=font, size=35)
plt.xlabel('The number of unrecognized samples', fontproperties=font, size=45, labelpad=20)
plt.show()
# fig.savefig(pjoin(support_path, 'unrecognized_samples.jpg'), dpi=300, bbox_inches='tight')
```
